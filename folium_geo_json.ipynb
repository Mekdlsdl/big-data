```python
import folium
import pandas as pd

geo_json = 'https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json'

raw_data = [
    ['강남구', 30],
    ['강동구', 30],
    ['강서구', 0],
    ['강북구', 30],
    ['관악구', 30],
    ['광진구', 30],
    ['구로구', 30],
    ['금천구', 30],
    ['노원구', 30],
    ['도봉구', 30],
    ['동대문구', 30],
    ['동작구', 30],
    ['마포구', 30],
    ['서대문구', 30],
    ['서초구', 30],
    ['성동구', 40],
    ['성북구', 30],
    ['송파구', 30],
    ['양천구', 30],
    ['영등포구', 30],
    ['용산구', 10],
    ['은평구', 30],
    ['종로구', 30],
    ['중구', 40],
    ['중랑구', 30]
]

raw_data2 = [
    ['강서구', 30],
    ['강남구', 0],
    ['강동구', 0],
    ['강북구', 0],
    ['관악구', 0],
    ['광진구', 0],
    ['구로구', 0],
    ['금천구', 0],
    ['노원구', 0],
    ['도봉구', 0],
    ['동대문구', 0],
    ['동작구', 0],
    ['마포구', 0],
    ['서대문구', 0],
    ['서초구', 0],
    ['성동구', 0],
    ['성북구', 0],
    ['송파구', 0],
    ['양천구', 0],
    ['영등포구', 0],
    ['용산구', 0],
    ['은평구', 0],
    ['종로구', 0],
    ['중구', 0],
    ['중랑구', 0]
]

data = pd.DataFrame(raw_data, columns=['name', 'value'])
data2 = pd.DataFrame(raw_data2, columns=['name', 'value'])

m = folium.Map(
    location=[37.566345, 126.977893],
    tiles='cartodbpositron'
)
m2 = folium.Map(
    location=[37.566345, 126.977893],
    tiles='cartodbpositron'
)
folium.Choropleth(
    geo_data=geo_json,
    name='choropleth',
    data=data2,
    columns=['name', 'value'],
    key_on='feature.properties.name',
    fill_color='Blues',
    fill_opacity=1,
    line_opacity=0.5
).add_to(m)

folium.Choropleth(
    geo_data=geo_json,
    name='choropleth',
    data=data,
    columns=['name', 'value'],
    key_on='feature.properties.name',
    fill_color='Reds',
    fill_opacity=0.6,
    line_opacity=0.5
).add_to(m)

m
```
