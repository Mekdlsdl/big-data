```python
import warnings
warnings.filterwarnings(action='ignore') 
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
def findBook(bookName):
    url = "https://data4library.kr/autoComplete?searchText="

    bookName = bookName.replace("[", "").replace("]", "")

    url = url + bookName

    res = requests.get(url, verify=False)
    soup = BeautifulSoup(res.content, 'html')

    title = soup.select_one("html > body > p")

    for a in title:
      # print(a[24:-14])
      if len(a[24:-14]) > 3:
        return ("책 있음")
      else:
        return ("책 없음")
```

```python
def bookOutput(url):
  res = requests.get(url)
  soup = BeautifulSoup(res.content, 'xml')

  resultset = soup.find_all("description")

  text = []

  for a in resultset:
    text.append(str(a))

  bookSplit = text[1].split("[도서]")

  bSplit = []

  for i in range(len(bookSplit)):
    bSplit.append(bookSplit[i].split("&lt;/b&gt;&lt;/a&gt;&lt;br&gt;"))

  infoSplit = []
  nameSplit = []
  for i in range(1,len(bSplit)):
    infoSplit.append(bSplit[i][1].split('|')[:2])
    nameSplit.append(bSplit[i][0])

  bookList = {}
  for i in range(len(infoSplit)):
    bookList[nameSplit[i]] = (infoSplit[i][0], infoSplit[i][1])

  titles = []
  for a in bSplit[0]:
    titles.append(a.split("연령별 관심도서"))

  titles2 = []
  for a in titles:
    titles2.append(a[1].split(" - "))

  title = []
  for a in titles2:
    title.append(a[1])
    title.append(a[2].split("&")[0])

  for a in title:
    print(a, end = " ")

  print("\n")
  for a in bookList:
    print(f"<{a} >\n저자: {bookList[a][0]}\n출판:{bookList[a][1]}")
    print(f"{findBook(a)}\n-------------")
```

```python
bookOutput('http://www.yes24.com/_par_/Rss/KAGEF001023.xml')
```
